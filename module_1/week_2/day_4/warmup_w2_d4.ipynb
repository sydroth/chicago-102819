Use Seaborn's lmplot or regplot to show a scatter plot and a line of best fit. Use "total_bill" as the x-axis and tip as your y-axis. Experiment with different hues and markers!

```python
tips['tip_percent']= round(((tips['tip'] / tips['total_bill'])* 100), 2)
```

```python
sns.lmplot('total_bill', 'tip', palette='helix', data=tips, scatter=True)

#helix=sns.color_palette("cubehelix")
#sns.stripplot(sns.color_palette("cubehelix"))
```

```python
sns.stripplot(x='total_bill', y='tip_percent', data=tips, jitter=True, palette="cubehelix", edgecolor='gray', linewidth=1)
```
